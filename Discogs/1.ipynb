# API DISCOGS

```python
import requests
import pandas as pd
import json
import time
import numpy as np
```

##Â Llamamos a la API 

```python
# Lista para almacenar datos de cada artista
artist_data = []
for artist_id in range(1, 30):
    url = f"https://api.discogs.com/artists/{artist_id}?callback=callbackname"
    response = requests.get(url)
    respuesta = response.text.replace("callbackname(", "")[:-1]
    artista = json.loads(respuesta)
    if 'data' in artista and 'name' in artista['data']: 
        data = artista['data']['name']
        artist_data.append({'artist_id': artist_id, 'name': data, })
    else:
        pass
    # Pausa para evitar sobrecargar la API (usualmente recomendado en APIs)
    time.sleep(1)  # Medio segundo de espera

# Crear un DataFrame con todos los datos de los artistas
df = pd.DataFrame(artist_data)
df.head(20)
```

```python
df.to_csv('Artist.csv', header=True, index= False)
```

## Introducimos id para vincular despues con los artistas, uri, albums en links y el perfil de los 24 artistas

```python
df = pd.read_csv("Artist.csv")
df
```

```python
ids = df.artist_id.unique()
ids
```

```python
id_list = []
uri_list = []
releases_list = []
profile_list=[]

for id_ in ids:
    url = f"https://api.discogs.com/artists/{id_}?callback=callbackname"
    response = requests.get(url)
    respuesta = response.text.replace("callbackname(", "")[:-1]
    artista = json.loads(respuesta)
    artista = artista['data']
    print(artista['id'])
    id_list.append(artista['id'])
    uri_list.append(artista['uri'])
    releases_list.append(artista['releases_url'])
    profile_list.append(artista['profile'])

    time.sleep(1)
```

```python
artistas_df = pd.DataFrame({"artist_id" : id_list, "uri":uri_list, "albums":releases_list, "descripcion":profile_list})
artistas_df
```

```python
artistas_df.to_csv("artistas_df.csv", header=True, index=False)
```

# Tenemos nuestra tabla para ir cogiendo datos de la api de albums

```python
import pandas as pd
import requests
import time
```

```python
artists_df = pd.read_csv("artistas_completo.csv")
artists_df
```

## Vemos los albumes del data 28 para ver si funciona.

```python
url = f"https://api.discogs.com/artists/28/releases?callback=callbackname"
response = requests.get(url)
respuesta = response.text.replace("callbackname(", "")[:-1]
releases = json.loads(respuesta)
test = releases['data']['releases']
test
```

```python
## Metemos los albums de cada artista y extraemos el id del album, titulo y aÃ±o formato.

lista_artists_ids = artists_df.artist_id.unique()
album_df = pd.DataFrame(columns = ["artist_id","album_id","album_name","album_year", "album_format"])
for artist_id in lista_artists_ids:
    lista_ids=[]
    lista_titulos=[]
    lista_aÃ±os=[]
    lista_stats=[]
    lista_format=[]
    url = f"https://api.discogs.com/artists/{artist_id}/releases?callback=callbackname"
    response = requests.get(url)
    respuesta = response.text.replace("callbackname(", "")[:-1]
    releases = json.loads(respuesta)
    lista_releases = releases['data']['releases']
    for release in lista_releases:
        if 'id' in release:
            lista_ids.append(release['id'])
        else: 
            lista_ids.append(0)
        if 'title' in release:
            lista_titulos.append(release['title'])
        else: 
            lista_titulos.append(0)
        if 'year' in release:
            lista_aÃ±os.append(release['year'])
        else: 
            lista_aÃ±os.append(0)
        if 'format' in release:
            lista_format.append(release['format'])
        else: 
            lista_format.append(0)
    aux_df = pd.DataFrame(columns = ["artist_id","album_id","album_name","album_year", "album_format"])
    aux_df["album_id"] = lista_ids
    aux_df["album_name"] = lista_titulos
    aux_df["album_year"] = lista_aÃ±os
    aux_df["artist_id"]= artist_id
    aux_df["album_format"]= lista_format
    album_df = pd.concat([album_df,aux_df])
    print(len(album_df))
    time.sleep(1)
```

```python
album_df
```

## Junto con merge los albums con sus artistas por el ID

```python
artistas = artists_df.merge(album_df, on= "artist_id", how='inner')
artistas
```
